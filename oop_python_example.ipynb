```python
class Empoly:
    
    # Constructor
    def __init__(self, name,jop,age,salary=0):
        self.name = name 
        self.jop = jop
        self.age = age
        self.salary=salary
    
  
    
    def emp_info(self):
        return "اسم الموظف :"+ self.name + "\n" +\
        " الوظيفة:" + self.jop + "\n" +\
        " العمر :"+ str(self.age)

        
    def increase_salary(self,amount):
        self.salary +=amount
        print ("مقدار الزيادة بالراتب :" , amount ,"\n", "الراتب بعد الزيادة :" ,self.salary)
    


emp=Empoly("ahmed","engineer",40,1000)


print(emp.emp_info())
emp.increase_salary(300)
```
